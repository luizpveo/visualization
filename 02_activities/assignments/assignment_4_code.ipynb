```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TCDSB - Schools - 4326.csv')
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# number of schools by level 
sns.countplot(x='TCDSB_LEVEL', data=df, palette='viridis', ax=ax1,order=df['TCDSB_LEVEL'].value_counts().index)
ax1.set_xlabel('Level')
ax1.set_ylabel('Number of Schools')
ax1.set_title('Number of Schools by Level')

# number of schools by Municipality 
sns.countplot(x='MUNICIPALITY', data=df, palette='viridis', ax=ax2, order=df['MUNICIPALITY'].value_counts().index)
ax2.set_xlabel('Number of Schools')
ax2.set_ylabel('Municipality')
ax2.set_title('Number of Schools by Municipality')

plt.tight_layout()
plt.show()
```
